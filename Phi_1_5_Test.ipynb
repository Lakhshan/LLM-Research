## Phi 1.5


````python
import torch
import transformers
from transformers import AutoTokenizer, AutoModelForCausalLM
````

````python
torch.set_default_device('cuda')
````

````python


model = AutoModelForCausalLM.from_pretrained("microsoft/phi-1_5",
                                             trust_remote_code=True,
                                             torch_dtype="auto")

tokenizer = AutoTokenizer.from_pretrained("microsoft/phi-1_5",
                                          trust_remote_code=True,
                                          torch_dtype="auto")

````

````python
def generate(input_text, max_length=1000):
    inputs = tokenizer(input_text, return_tensors="pt", return_attention_mask=False)
    outputs = model.generate(**inputs, max_length=max_length)
    text = tokenizer.batch_decode(outputs)[0]
    print(text)
````

## Instruction Answering

````python
generate('Which animal is better as a pet cat or a dog ?.\n\n Answer:',
         max_length=128)
````

````python
generate('How many ways can the letters in the word COMMON be arranged ?\n\n Answer:',
         max_length=128)
````

````python
generate('Any tip to sleep fast ?.\n\n Answer:',
         max_length=128)
````

````python
generate('What is the average time for the sun light to reach the earth ?.\n\n Answer:',
         max_length=128)
````
